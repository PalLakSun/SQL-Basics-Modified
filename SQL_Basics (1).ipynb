9.What is the Difference Between a Primary Key and a Foreign Key?

Ans-

Primary Key (PK)

Purpose: Uniquely identifies each record in a table.

Characteristics:

-Must be unique for every row.

-Cannot be NULL.

-There can be only one primary key per table (can be composite though, using multiple columns)

Example:
In an Employees table:

EmployeeID (PK)

Each employee has a unique EmployeeID.

Foreign Key (FK)

Purpose: Establishes a link between two tables. It refers to the primary key in another table.

Characteristics:

-Can have duplicate values.

-Can be NULL (if optional relationship).

-There can be multiple foreign keys in a table.

Example:
In a Projects table:

EmployeeID (FK)

This EmployeeID refers to the primary key of the Employees table, showing which employee is assigned to the project.

In short,

Primary Key = unique ID for a record and defines identity within table

Foreign Key = reference to a primary key in another table and defines relationship between tables


10.What Are Constraints in SQL and Why Are They Used?

Ans-Constraints are rules applied to table columns to enforce data integrity and consistency.

-They restrict the type of data that can be inserted or updated in a table.

Why Are Constraints Used?

-Ensure Accuracy and Reliability

-Prevents invalid data from entering the database.

-Maintain Data Integrity

-Keeps relationships between tables consistent.

-Control Business Rules

-Enforces rules like “salary cannot be negative” or “each employee must have a unique ID”.

Common Types of Constraints

1.PRIMARY KEY
Uniquely identifies each row in a table. Cannot be NULL.

2.FOREIGN KEY
Ensures that values in a column match values in another table (maintains referential integrity).

3.UNIQUE
Ensures all values in a column are unique (can be NULL).

4.NOT NULL
Ensures a column cannot have NULL values.

5.CHECK
Ensures that column values satisfy a condition (e.g., Salary > 0).

6.DEFAULT
Provides a default value if none is specified during insertion.
